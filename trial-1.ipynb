```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
reviews = pd.read_csv("reviews_short.txt")
df = reviews[['user_rating', 'user_review']]
df
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
df['neg'] = df['user_review'].apply(lambda x:sia.polarity_scores(x)['neg'])
df['neu'] = df['user_review'].apply(lambda x:sia.polarity_scores(x)['neu'])
df['pos'] = df['user_review'].apply(lambda x:sia.polarity_scores(x)['pos'])
df['compound'] = df['user_review'].apply(lambda x:sia.polarity_scores(x)['compound'])
```

```python
df
```

```python
pos_review = [j for i,j in enumerate(df['user_review']) if df['compound'][i] > 0.2]
neu_review = [j for i,j in enumerate(df['user_review']) if 0.3>=df['compound'][i]>=-0.2]
neg_review = [j for i,j in enumerate(df['user_review']) if df['compound'][i] < -0.2]
```

```python
print("Percentage of positive review: {}%".format(len(pos_review)*100/len(df['user_review'])))
print("Percentage of neutral review: {}%".format(len(neu_review)*100/len(df['user_review'])))
print("Percentage of negative review: {}%".format(len(neg_review)*100/len(df['user_review'])))
```

```python
# Function to normalise hotel ratings in the range of -1 to 1

def normalise_ratings(r):
    if(r==5):
        return 1
    if(r==4):
        return 0.5
    if(r==3):
        return 0
    if(r==2):
        return -0.5
    if(r==1):
        return -1
```

```python
df['normalised'] = df['user_rating'].apply(lambda x:normalise_ratings(x))
df
```
